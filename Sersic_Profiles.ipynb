```python
import numpy as np
import matplotlib.pyplot as plt

def sersic_profile(R, n, I0=1.0, R0=5.0):
    return I0 * np.exp(- (R / R0)**(1.0 / n))

#Radius
R = np.linspace(0, 50, 500)

#n-indices
n_values = [0.1, 0.5, 1.0, 2.0, 5.0, 10.0]

plt.figure(figsize=(8, 5))

for n in n_values:
    plt.plot(R, sersic_profile(R, n), label=f'n = {n}')

plt.xlabel('Radius $R$ (kpc)', fontsize=12)
plt.ylabel('Normalized Surface Brightness $I(R) / I_0$', fontsize=12)
plt.title('Sersic Profiles for Different Sersic Indices', fontsize=14)
plt.legend(title='Sersic index', frameon=False)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(0, 50)
plt.ylim(0, 1.05)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def bn_approx(n):
    return (2*n - 1/3
            + 4/(405*n)
            + 46/(25515*n**2)
            + 131/(1148175*n**3)
            - 2194697/(30690717750*n**4))

def sersic_half_light(R, n, Ie=1.0, Re=5.0):
    b = bn_approx(n)
    return Ie * np.exp(-b * ((R / Re)**(1.0 / n) - 1))

R = np.linspace(0, 50, 500)
n_values = [0.5, 1.0, 2.0, 5.0, 10.0]

plt.figure(figsize=(8,5))
for n in n_values:
    plt.plot(R, sersic_half_light(R, n), label=f'n = {n}')

plt.xlabel('Radius $R$ (kpc)', fontsize=12)
plt.ylabel(r'Normalized Surface Brightness $I(R)/I_e$', fontsize=12)
plt.title('Sersic Profiles (from falf-light eqn)', fontsize=14)
plt.legend(title='Sersic index', frameon=False)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(0,50)
plt.ylim(0,1.05)

plt.tight_layout()
plt.show()
```
